# **Pilhas**
Uma pilha é uma estrutura que armazena dados como olha pulha de pratos. Você pode colocar um prato no topo da pilha e, quando precisar usar um, pegará o que estiver no topo. Isso porque o último prato que foi adicionado será o primeiro a ser retirado da pulha.

Uma pulha é uma estrutura de dados que armazena o dado em uma ordem específica, com algumas restrições:

* Elementos de dados só podem ser inseridos no fim da pilha (*push*)
* Elementos de dados só podem ser excluídos do final da pilha (*pop*)
* Apenas o último elemento de dados pode ser lido na pilha (*peek*)


image.png

Uma pilha (*stack*) permite armazenar e ler dados em uma extremidade, e o elemento adicionado por último é o primeiro a ser selecionado. Por isso, a pilha é conhecida como uma estrutura LIFO (*last in, first out* - o último a entrar é o primeiro a sair).

# **Operações com pilhas**

* Quando um elemento é adicionado ao topo da pulha, essa operação chama-se *push*;
* Quando um elemento é selecionado (ou seja, removido) no topo, a operação é chamada de *pop*;
* Existe uma operação denominada *peek*, que é quando o elemento do topo é visualizado sem ser removido da pilha.

As operações com pilhas usam como referência um ponteiro, normalmente denominado *top*, que indica o topo da pilha.



Geralmente, as pilhas podem ser implementadas com o uso de arrays e listas encadeadas. As implementações baseadas em arrays terão tamanhos fixos para as pilhas, enquanto as baseadas em listas encadeadas podem ter pilhas de tamanhos variáveis.

# **Implementação com arrays**
No caso da implementação de uma pilha baseada em array (na qual a pilha tem tamanho fixo), é importante verificar se a pilha está ou não cheia, já que tentar inserir um elemento em uma pilha cheia gerará um erro, chamado *overflow* (estouro de pilha). Da mesma forma, tentar aplicar uma operação pop a uma pilha vazia causa um erro conhecido como *underflow*.


image.png

O código a seguir exemplifica a implementação de pilhas usando arrays.


```python
class StackArray:
    def __init__(self, capacity):
        self.capacity = capacity  # Capacidade máxima da pilha
        self.stack = [None] * capacity  # Inicializa o array com capacidade definida
        self.top = -1  # Inicializa o índice do topo da pilha

    def is_empty(self):
        return self.top == -1  # Verifica se a pilha está vazia

    def is_full(self):
        return self.top == self.capacity - 1  # Verifica se a pilha está cheia

    def push(self, item):
        if self.is_full():
            print("Erro: Stack overflow")
        else:
            self.top += 1
            self.stack[self.top] = item  # Adiciona o elemento ao topo da pilha

    def pop(self):
        if self.is_empty():
            print("Erro: Stack underflow.")
            return None
        else:
            item = self.stack[self.top]  # Armazena o elemento do topo
            self.top -= 1  # Atualiza o topo para o próximo elemento
            return item

    def peek(self):
        if self.is_empty():
            print("Erro: Pilha vazia.")
            return None
        else:
            return self.stack[self.top]  # Retorna o elemento do topo da pilha

    def display(self):
        if self.is_empty():
            print("A pilha está vazia.")
        else:
            print("Elementos na pilha:")
            for i in range(self.top, -1, -1):
                print(self.stack[i])
```

O método `__init__` inicializa a pilha com uma capacidade máxima definida. `self.capacity` armazena o número máximo de elementos que a pilha pode conter. `self.stack` é um array inicializado com None que armazenará os elementos da pilha. `self.top` é o índice que aponta para o topo da pilha.

O método `is_empity()` verifica se a pilha está vazia. Ele retorna *True* se o índice do topo for igual a -1, indicando que a pilha está vazia, e *False* caso contrário.

O método `is_full()` verifica se a pilha está cheia. Ele retorna *True* se o índice do topo for igual à capacidade máxima menos um, indicando que a pilha está cheia, e *False* caso contrário.

O método `push()` adiciona um elemento à pilha. Se a pilha estiver cheia, exibe uma mensagem de erro. Caso contrário, incrementa o índice do topo e adiciona o elemento ao array `self.stack` na posição do topo.

O método `pop()` remove o elemento do topo da pilha e o retorna. Se a pilha estiver vazia, exibe uma mensagem de erro e retorna *None*. Caso contrário, armazena o elemento do topo em item, decrementa o índice do topo e retorna o item.

O método `peek()` retorna o elemento do topo da pilha sem removê-lo. Se a pilha estiver vazia, exibe uma mensagem de erro e retorna *None*. Caso contrário, retorna o elemento do topo.

O método `display()` exibe todos os elementos da pilha. Se a pilha estiver vazia, exibe uma mensagem indicando que está vazia. Caso contrário, itera pelo array `self.stack` do topo até o início, imprimindo cada elemento da pilha.

Abaixo estão exemplos de uso da pilha, incluindo situações de *overflow* e *underflow*.

```python
stack1 = StackArray(2)  # Criando uma pilha com capacidade para 2 elementos

#Adicionando elementos à pilha
stack1.push(1)
stack1.push(2)

# Exemplo de overflow (tentando adicionar mais elementos que a capacidade)
stack1.push(3)

print("\n")

# Esvaziando a lista removendo os dois elementos
stack1.pop()
stack1.pop() # <-- Lista vazia

#Tentando remover elemento de lista vazia
stack1.pop()
```

# **Implementação com listas encadeadas**
Para implementar uma pilha com uma lista encadeada, são necessários dois elementos principais: os nós da lista encadeada e a própria pilha que gerencia esses nós. Cada nó armazena um elemento da pilha e uma referência para o próximo nó na pilha.

O código a seguir implementa pilha com o uso de uma lista encadeada.



```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def is_empty(self):
        return self.top is None

    def push(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.size += 1

    def pop(self):
        if self.is_empty():
            print("Erro: Pilha vazia.")
            return None
        else:
            data = self.top.data
            self.top = self.top.next
            self.size -= 1
            return data

    def peek(self):
        if self.is_empty():
            print("Erro: Pilha vazia.")
            return None
        else:
            return self.top.data

    def display(self):
        current = self.top
        if self.is_empty():
            print("Pilha vazia.")
        else:
            print("Elementos na pilha:")
            while current is not None:
                print(current.data)
                current = current.next

stack = Stack()

stack.push("Pão")
stack.push("Queijo")
stack.push("Presunto")

stack.display()

print("Elemento no topo da pilha:", stack.peek())

#Removendo os elementos da pilha
while not stack.is_empty():
    stack.pop()

```

A operação `push` é uma operação importante executada em uma pilha; ela é usada para adicionar um elemento no topo da pilha. Para adicionar um novo nó à pilha, primeiro verificamos se a pilha já tem alguns itens ou se está vazia.

No caso em questão, não precisamos verificar a condição de *overflow* porque não é necessário corrigir o tamanho da pilha, como ocorre na implementação com o uso de arrays. Se a pilha já tiver alguns elementos, precisaremos fazer duas coisas:
1. O novo nó deve estar com seu ponteiro next apontando para o nó que anteriormente estava no topo.
2. Devemos inserir esse novo nó no topo da pilha apontando `self.top` para o nó recém-adicionado.

Se a pilha estiver vazia e o nó a ser adicionado for o primeiro elemento, precisamos transformar este no elemento do topo. Ou seja, `self.top` apontará para esse novo nó.

---

Outra operação importante aplicada à pilha é a `pop`. Nessa operação, o elemento do topo da pilha é lido e, em seguida, removido da pilha. O método pop retorna o elemento do topo da pilha, mas retorna None se a pilha estiver vazia.

Para implementar a operação pop em uma pilha, temos de fazer o seguinte:
1. Primeiro, verificamos se a pilha está vazia. A operação pop não é permitida em uma pilha vazia.
2. Se a pilha não estiver vazia, verificaremos se o nó do topo está com seu atributo `next` apontando para algum outro nó. Se estiver, isso significa que a pilha contém elementos e o nó do topo está apontando para o próximo nó da pilha. Para aplicar a operação `pop`, precisamos alterar o ponteiro top. O próximo nó deve ser o do topo. Faremos isso apontando `self.top` para `self.top.next`.
3. Quando só houver um nó na pilha, ela ficará vazia após a execução da operação `pop`. Precisamos alterar o ponteiro `top` para apontar para *None*.
4. Com a remoção desse nó, `self.top` apontará para *None*.
5. Também decrementaremos o tamanho da pilha em 1 unidade se ela não estiver vazia.

---

O método `peek()` retorna o elemento do topo da pilha sem excluí-lo. A única diferença entre peek e pop
é que o método peek apenas retorna o elemento do topo; no entanto, no caso de um método
pop, o elemento do topo é retornado, mas ele também é excluído da pilha.

A operação `peek` permite examinar o elemento do topo sem alterar a pilha. Essa operação
é muito simples. Se houver um elemento no topo, ela retornará seus dados; caso contrário,
retornará None (logo, o comportamento de peek é semelhante ao de `pop`).

---

As operações push, pop e peek têm complexidade de tempo constante (O(1)):
* Push (Adicionar um elemento à pilha): Adicionar um elemento ao final de uma lista tem complexidade de tempo constante, já que não é necessário realocar memória ou reorganizar os elementos.

* Pop (Remover um elemento da pilha): Remover o último elemento de uma lista também tem complexidade de tempo constante, pois a operação de pop não requer reorganização de elementos na lista.

* Peek (Visualizar o elemento do topo sem removê-lo): Obter o último elemento de uma lista também é uma operação de complexidade constante, já que podemos acessar diretamente o último elemento usando o índice -1.

# **Aplicações com pilhas**

Abaixo estão algumas das principais razões pelas quais as pilhas são importantes:

0. Gestão de Memória: As pilhas são amplamente utilizadas em linguagens de programação para gerenciar a memória de forma eficiente. Por exemplo, em linguagens como C e C++, as variáveis locais e parâmetros de função são armazenados na pilha de execução do programa. Isso permite um uso eficiente da memória, já que os recursos são liberados automaticamente quando uma função é encerrada.

1. Implementação de Algoritmos: Muitos algoritmos e estruturas de dados são baseados em pilhas. Por exemplo, a resolução de expressões matemáticas em notação polonesa reversa (RPN) ou a avaliação de expressões infixas usando pilhas para o gerenciamento dos operadores e operandos.
Outros algoritmos como busca em profundidade (depth-first search) em grafos ou inversão de sequências também fazem uso de pilhas de forma eficiente.

2. Sistema de Undo (Desfazer): Pilhas são frequentemente utilizadas em sistemas que oferecem a funcionalidade de desfazer (undo) operações. Cada ação realizada pelo usuário é adicionada à pilha e pode ser desfeita removendo-se elementos da pilha na ordem reversa.

3. Navegação em Histórico: Navegadores da web utilizam pilhas para gerenciar o histórico de páginas visitadas. Ao navegar para uma nova página, ela é adicionada à pilha, e ao pressionar o botão "Voltar", a página é removida da pilha, permitindo a navegação reversa.

4. Análise Sintática: Em compiladores e analisadores sintáticos, as pilhas são usadas para implementar a análise sintática descendente (bottom-up parsing) e para rastrear a estrutura da gramática de uma linguagem de programação.

5. Empilhamento de Chamadas (Call Stack): Em linguagens de programação, as chamadas de função são gerenciadas por uma pilha de execução (call stack). Isso permite que as funções sejam chamadas recursivamente e que o fluxo de execução do programa seja controlado de forma organizada.

6. Backtracking e Recursão: Algoritmos de backtracking e funções recursivas frequentemente usam pilhas para armazenar o estado atual do processo. Cada chamada recursiva ou etapa de backtracking é representada como um elemento na pilha.

Exemplos de aplicações específicas das pilhas incluem:

* Processamento de fórmulas matemáticas.
* Implementação de sistemas de gerenciamento de texto (undo/redo).
* Navegação em estruturas de diretórios de um sistema operacional.
* Validação de expressões regulares.
* Resolução de quebra-cabeças (como o jogo "Torre de Hanoi").
* Avaliação de expressões lógicas.
* Implementação de algoritmos de ordenação como QuickSort usando pilha de execução.
